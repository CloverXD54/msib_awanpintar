```python
number = input("ketiklah satu angka, kami akan memberi keterangan ganjil atau genap: ")
number = int(number)
if number % 2 == 0:
  print(f"\nAngka {number} yang anda ketik adalah genap.")
else:
  print(f"\nAngka {number} yang anda ketik adalah ganjil.")
```
